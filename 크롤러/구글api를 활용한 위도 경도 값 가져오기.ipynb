```python
import googlemaps
import pandas as pd
from tqdm import tqdm

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/Colab Notebooks/서울교통공사_2022년.csv'

# CSV 파일 읽기
df_shake = pd.read_csv(csv_file_path, encoding='cp949')

# API키 입력
mykey = "AIzaSyDqIXXOGU4_KtjJk_50ZetSL6p-GzYSPeU"
maps = googlemaps.Client(key=mykey)  # my key값 입력

# 역명에서 괄호와 그 안의 내용 제거
df_shake['역명'] = df_shake['역명'].apply(lambda x: x.split('(')[0].strip())

# '역' 추가 (없을 경우에만 추가)
df_shake['역명'] = df_shake['역명'].apply(lambda x: x + '역' if not x.endswith('역') else x)

# 위도, 경도 변환하는 함수 생성
def trans_geo(역명):
    try:
        geo_location = maps.geocode(역명)[0].get('geometry')
        lat = geo_location['location']['lat']
        lng = geo_location['location']['lng']

        return [lat, lng]
    except:
        return [0, 0]

# 실행
for idx, row in tqdm(df_shake.iterrows(), total=len(df_shake)):
    역명 = row['역명']
    lat, lng = trans_geo(역명)
    df_shake.at[idx, 'latitude'] = lat
    df_shake.at[idx, 'longitude'] = lng

# 수정된 DataFrame을 새로운 CSV 파일로 저장
new_csv_path = '/content/drive/MyDrive/Colab Notebooks/서울교통공사_2022년_with_geo.csv'
df_shake.to_csv(new_csv_path, encoding='cp949', index=False)
```
